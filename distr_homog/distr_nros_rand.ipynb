# Distribucion homogena de particulas en x, y, z

```python
n = 1000
p = np.ndarray([n,4])



for i in range(n):
    
    x = rm.random()
    y = rm.random()
    z = rm.random()
    
    p[i,:] = [i,x,y,z]
    
#print p
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

ax1.plot(p[:,1],p[:,2],'o')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.plot(p[:,1],p[:,3],'o')
ax2.set_xlabel('x')
ax2.set_ylabel('z')
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(p[:,1], p[:,2], p[:,3], marker='o')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

# Distribucion homogena de particulas en r

```python
n = 1000
p = np.ndarray([n,4])

for i in range(n):
    
    r3 = rm.uniform(0, 1.)
    ct = rm.uniform(-1, 1)
    fi = rm.uniform(0, 2*np.pi)
    
    st = np.sqrt(1 - ct**2)
        
    x = (r3*3)**(1/3.) * st * np.cos(fi)
    y = (r3*3)**(1/3.) * st * np.sin(fi)
    z = (r3*3)**(1/3.) * ct
    
    p[i,:] = [i,x,y,z]
    
#print p
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5), tight_layout=True)

ax1.plot(p[:,1],p[:,2],'o')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.plot(p[:,1],p[:,3],'o')
ax2.set_xlabel('x')
ax2.set_ylabel('z')
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(p[:,1], p[:,2], p[:,3], marker='o')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

# Esfera con distribucin r**-2

```python
n = 1000
p = np.ndarray([n,4])

for i in range(n):
    
    r = rm.uniform(0, 1.)
    ct = rm.uniform(-1, 1)
    fi = rm.uniform(0, 2*np.pi)
    
    st = np.sqrt(1 - ct**2)
        
    x = r * st * np.cos(fi)
    y = r * st * np.sin(fi)
    z = r * ct
    
    p[i,:] = [i,x,y,z]
    
#print p
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5), tight_layout=True)

ax1.plot(p[:,1],p[:,2],'o')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.plot(p[:,1],p[:,3],'o')
ax2.set_xlabel('x')
ax2.set_ylabel('z')
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(p[:,1], p[:,2], p[:,3], marker='o')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
